### **Data Cleaning**

```python
# For checking missing values
print(train_data.isnull().sum())
```

```python
# For checking missing values
print(test_data.isnull().sum())
```

```python
# Handle missing values using median
train_data["Age"] = train_data["Age"].fillna(train_data["Age"].median())
train_data.fillna("None", inplace=True)
train_data
```

```python
test_data["Age"] = test_data["Age"].fillna(test_data["Age"].median())
test_data.fillna("None", inplace=True)
test_data
```

### **Mapping Categorical values into Numerical values**

```python
train_data['Sex'] = train_data['Sex'].map({'male': 0, 'female': 1})
train_data
```

```python
test_data['Sex'] = test_data['Sex'].map({'male': 0, 'female': 1})
test_data
```

```python
train_data['Embarked'] = train_data['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})
train_data
```

```python
test_data['Embarked'] = test_data['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})
test_data
```

```python
# Drop columns which will not be used
train_data.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1, inplace=True)
train_data
```

```python
test_data.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1, inplace=True)
test_data
```

### **Exploratory Data Analysis (EDA)**

```python
sns.pairplot(train_data, hue='Survived')
plt.show()
```

```python
sns.pairplot(test_data)
plt.show()
```

```python
sns.countplot(x='Survived', data=train_data, color="blue")
plt.title('Survival Count')
plt.show()
```

```python
sns.countplot(x="Sex", hue="Survived", data=train_data,palette = "Set1", edgecolor="black")
plt.show()
```

```python
sns.countplot(x='Pclass', hue='Survived', data=train_data, palette = "Set2", edgecolor="black")
plt.title('Pclass: Survived vs Not Survived')
plt.show()
```

```python
sns.boxplot(x='Survived', y='Age', data=train_data, color="green")
plt.title('Age vs. Survival')
plt.show()
```

```python
sns.boxplot(x='Pclass', y='Age', data=test_data, color="magenta")
plt.title('Age vs. Pclass')
plt.show()
```

```python
plt.scatter(train_data['Age'], train_data['Fare'])
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Age vs. Fare')
plt.show()
```

```python
plt.scatter(test_data['Age'], test_data['Pclass'])
plt.xlabel('Age')
plt.ylabel('Pclass')
plt.title('Age vs. Pclass')
plt.show()
```

```python
sns.violinplot(x ="Pclass", y ="Age", hue ="Survived", data = train_data, split = True, palette="dark:red")
```
